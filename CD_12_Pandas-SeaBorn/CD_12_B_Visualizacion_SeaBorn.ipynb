# Análisis exploratorio de datos (EDA)

El análisis exploratorio de datos o Exploratory Data Analysis(EDA) es un enfoque para analizar conjuntos de datos para resumir sus principales características. Se utiliza para comprender los datos, obtener un contexto al respecto, comprender las variables y las relaciones entre ellas y formular hipótesis que podrían ser útiles a la hora de construir modelos predictivos.

Al visualizar de forma eficaz las variables de un conjunto de datos y sus relaciones,  un científico de datos puede comprender rápidamente tendencias, valores atípicos y patrones. Esta comprensión se puede utilizar para contar una historia, impulsar decisiones y crear modelos predictivos.

**En esta lección, aprenderemos cómo realizar EDA utilizando la visualización de datos. Específicamente, nos centraremos en funciones asociadas a Seaborn para graficar.**

### ¿Por qué Seaborn?
_**Seaborn**_ es más que una mera librería de visualización, es una librería para representación estadística, ya que muestra fácilmente la relación que guardan los datos para detectar tendencias y patrones.

* _Manejo de los datos_: Seaborn es capaz de “entender” directamente un DataFrame, representando fácilmente distribuciones de datos o agregaciones, sin desarrollar muchas líneas de código.
* _Visualizaciones disponibles_: Los gráficos ayudan a entender los datos y para ello, es necesario seleccionar el gráfico más adecuado. La galería de seaborn es una de las más amplias, desarrollada para representar análisis estadísticos de forma sencilla.
* _Personalización_: El aspecto visual de un gráfico que tan laborioso es con matplotlib, se desarrolla de forma muy sencilla con seaborn.


#  2: Exploración grafica con Seaborn


<div class="alert alert-success">
Mostrar con una grafica de barras el numero total de datos por variedad
</div>

```python
sns.countplot(x = 'variety',  data = iris)
#x=campo, data=de donde vienen los datos
```

<div class="alert alert-success">

Mostrar con graficas de líneas los valores para de las variables numéricas
</div>

```python
fig, ax = plt.subplots(figsize=(8,7)) #Esta linea se agrega para configurar el tamaño de la grafica
sns.lineplot(data=iris , linewidth=2.5  )
```

```python
#Graficar los valores de sepal_length
sns.lineplot(data=iris['sepal_length'] , linewidth=2.5  )
```

<div class="alert alert-success">
Mostrar con histogramas los valores de las variables numéricas 
</div>

```python
sns.histplot(data=iris, element="step")
```

```python
#Histograma de petal_width
sns.histplot(data=iris, x="petal_width",element="step")
```

```python
#Histograma de petal_width separado por variedad
sns.histplot(data=iris, x="petal_width", element="step", hue="variety")
```

```python
#Histograma 2D de petal_width y sepal_width, separados por la variedad
fig, ax = plt.subplots(figsize=(8,7)) #Esta linea se agrega para configurar el tamaño de la grafica
sns.histplot(data=iris, x="petal_width", y= 'sepal_width',hue="variety")
```

<div class="alert alert-success">
Mostrar con diagramas de caja y bigotes los valores de las variables numéricas 
</div>

```python
sns.catplot(kind = 'box', data = iris)
```

<div class="alert alert-success">
Mostrar un diagrama de dispersión los valores de sepal_width y petal_width 
</div>

```python
sns.jointplot(data=iris, x = 'sepal_width', y = 'petal_width' )
```

```python
#Diagrama de dispersión separando por variedad
sns.jointplot(data=iris, x = 'sepal_width', y = 'petal_width',  hue="variety" )
```

```python
#Diagrama de dispersión separando por variedad
sns.jointplot(data=iris, x = 'sepal_width', y = 'petal_width' , kind="kde" ,  hue="variety")
#https://seaborn.pydata.org/generated/seaborn.jointplot.html
```

<div class="alert alert-success">
Mostrar un diagrama de dispersión de todos los valores numéricos 
</div>

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

```python
sns.pairplot( iris, kind = 'reg' )
```

```python
#columnas con datos numéricos
#iris_num=iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
iris_num = iris.select_dtypes(include=np.number)
iris_num
```

```python
# A continuación se mostrará la correlación que hay entre las variables númericas
plt.figure(figsize=(5,5),  dpi = 150)
corrMatrix = iris_num.corr().round(3)
sns.set( font_scale= 1.1 )
sns.heatmap( corrMatrix, annot = True , cmap = 'BrBG', center=0 )
plt.xticks( rotation = 90 )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
```

```python
# --> Contructor de consultas para obtener el token
#Token:
#c7077def-be35-af3e-cee9-769d0a5f2ae2
```

```python
#Población total de hombres serie histórica
#Llamado al API
url='https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000002/es/0700/false/BISE/2.0/c7077def-be35-af3e-cee9-769d0a5f2ae2?type=json'
respuesta = requests.get(url)
respuesta.status_code
```

```python
url='https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000002/es/0700/false/BISE/2.0/c7077def-be35-af3e-cee9-769d0a5f2ae2?type=json'
respuesta = requests.get(url)
if respuesta.status_code == 200:
    #print(respuesta.content) <--- formato json
    contenido = json.loads(respuesta.content) #convirtiendo formato json a diccionario

contenido
```

```python
len(contenido) #2 etiquetas
```

```python
contenido['Series'][0] #acceder al primer elemento
```

```python
contenido['Series'][0]['OBSERVATIONS'] #lista que contiene diccionarios
```

```python
#iterando en la lista de diccionarios, tomando año y valor de población

datos_hombres = contenido['Series'][0]['OBSERVATIONS']

#creando listas
poblacion_hombres=[]
periodo=[]

#iteración
for dato in datos_hombres:
    #agregando los valores de población a la lista poblacion_hombres
    #convirtiendo str a int, primero a float (".") luego a int
    poblacion_hombres.append(int(float(dato['OBS_VALUE'])))
    #agregando años a la lista periodo
    periodo.append(int(dato['TIME_PERIOD']))

print(poblacion_hombres)
print(periodo)
```

```python
#Para datos 'Mujeres'
url='https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000003/es/0700/false/BISE/2.0/c7077def-be35-af3e-cee9-769d0a5f2ae2?type=json'
respuesta = requests.get(url)
if respuesta.status_code == 200:
    #print(respuesta.content) <--- formato json
    contenido = json.loads(respuesta.content) #convirtiendo formato json a diccionario

datos_mujeres = contenido['Series'][0]['OBSERVATIONS']

#creando listas
poblacion_mujeres=[]
periodo=[]

#iteración
for dato in datos_mujeres:
    #agregando los valores de población a la lista poblacion_hombres
    #convirtiendo str a int, (primero a float (".") luego a int)
    poblacion_mujeres.append(int(float(dato['OBS_VALUE'])))
    #agregando años a la lista periodo
    periodo.append(int(dato['TIME_PERIOD']))

print(poblacion_mujeres)
print(periodo)
  
```

```python
#Creando DataFrame
df_poblacion = pd.DataFrame()
df_poblacion['periodo'] = periodo
df_poblacion['hombres'] = poblacion_hombres
df_poblacion['mujeres'] = poblacion_mujeres
df_poblacion['total'] = df_poblacion['hombres'] + df_poblacion['mujeres']
df_poblacion
```

```python
#Creando una gráfica del crecimiento de población
plt.plot(df_poblacion['periodo'], df_poblacion['total'], 'k-', label='Total')
plt.plot(df_poblacion['periodo'], df_poblacion['mujeres'], 'r:', label='Mujeres')
plt.plot(df_poblacion['periodo'], df_poblacion['hombres'], 'b-.', label='Hombres')
plt.legend()
plt.show()
```

```python
#Nuevo DataFrame
df = pd.melt(df_poblacion, id_vars='periodo', var_name='Categoria', value_name='Poblacion')
df
```

```python
import seaborn as sns
plt.figure(figsize=(15,8))
plt.title('Population growth in Mexico')
sns.barplot(x='periodo', y='Poblacion', data=df, hue='Categoria', palette=["#0B50D3","#66053A","#D30B57"])
plt.gca().legend().set_title('Gender')
plt.xticks(rotation=90)
plt.ylabel("Population (in hundreds of millions)")
plt.show()
```

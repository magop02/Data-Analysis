```python
#hectareas dato[-5]

with open('datos/BD_IncendiosSNIF_2021.csv') as archivo_csv:
    next(archivo_csv)
    datos_csv = csv.reader(archivo_csv)

    hectareas=[]
    for dato in datos_csv:
        hectareas.append(float(dato[-5]))

print("Total de hectáreas afectadas:", sum(hectareas))

#print(hectareas)
```

 <ul>
  <li>2. ¿En donde fue (estado y municipio) el incendio con mayor número de hectáreas afectadas y cuantas fueron  ?</li>

</ul> 

</div>

```python
#Incendio --> dato[1]
#Hectáreas --> dato[-5]
#Estado --> dato[11]
#Municipio --> dato[12]

with open('datos/BD_IncendiosSNIF_2021.csv') as archivo_csv:
    next(archivo_csv)
    datos_csv = csv.reader(archivo_csv)

    max_hectareas=0
    incendio={}
    for dato in datos_csv:
        hectareas=float(dato[-5])

        if hectareas > max_hectareas:
            max_hectareas = hectareas
            incendio = {'Clave de incendio':dato[1], 'Estado':dato[11], 'Municipio':dato[12], 'Hectareas afectadas':hectareas}

print(incendio)
```

```python
#Tipo de incendio --> dato[23]

with open('datos/BD_IncendiosSNIF_2021.csv') as archivo_csv:
    next(archivo_csv)
    datos_csv = csv.reader(archivo_csv)
    tipo_incendio={}
    for dato in datos_csv:
        tipo_incendio[dato[23]] = tipo_incendio.get(dato[23], 0) + 1
print(tipo_incendio)
```

```python
#Estado --> dato[11]
#Municipio --> dato[12]

with open('datos/BD_IncendiosSNIF_2021.csv') as archivo_csv:
    next(archivo_csv)
    datos_csv = csv.reader(archivo_csv)
    incendios_yuc={}
    for dato in datos_csv:
        if dato[11] == 'Yucatán':
            incendios_yuc[dato[12]] = incendios_yuc.get(dato[12], 0) + 1

print(incendios_yuc)
            
```

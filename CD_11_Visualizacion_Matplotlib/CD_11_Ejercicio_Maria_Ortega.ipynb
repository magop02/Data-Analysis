# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df= pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df.head()
```

```python
df=df.loc[:,'nombre':]
df = df.set_index('nombre').T
df=df.reset_index()
df = df.rename(columns = {'index': 'Fecha'})
df['Fecha'] = pd.to_datetime(df['Fecha'], format= '%d-%m-%Y')
df = df.set_index('Fecha')
```

```python
df.head()
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
total_nacional = df.sum(axis=1).resample('M').sum()
total_nacional
```

```python
#Estilo Matlab

plt.figure(figsize=(12,6))
plt.bar(total_nacional.index.strftime('%B'), total_nacional, color = 'c')
plt.title('Total nacional pacientes confirmados Covid',fontsize=10)
plt.xlabel('Meses',fontsize=10)
plt.ylabel('Total nacional',fontsize=10)

plt.show()
```

```python
#Estilo orientado a objetos

fig, ax = plt.subplots(figsize=(12,6), dpi=100)
ax.bar(total_nacional.index.strftime('%B'), total_nacional,  color = 'b')          
ax.set(title='Total nacional pacientes confirmados Covid', xlabel= 'Meses' , ylabel='Total nacional')

plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
df_seg = df[['DISTRITO FEDERAL', 'JALISCO', 'YUCATAN']]
df_seg = df_seg[df_seg.index.month.isin([8, 9, 10, 11])]
df_seg.head()
```

```python
val_df = df_seg['DISTRITO FEDERAL'].resample('M').sum()
val_jal = df_seg['JALISCO'].resample('M').sum()
val_yuc = df_seg['YUCATAN'].resample('M').sum()
```

```python
meses = ['Agosto', 'Septiembre', 'Octubre', 'Noviembre']
x_valores = np.arange(len(meses))
```

```python
#Estilo Matlab

plt.figure(figsize=(8,5))  
plt.bar(x_valores, val_df, width = 0.2 ,color = 'b',  label='DF')
plt.bar(x_valores + 0.25, val_jal, width = 0.2, color = 'r',  label='JAL')
plt.bar(x_valores + 0.5, val_yuc, width = 0.2, color = 'c',  label='YUC')

plt.xticks( x_valores , meses )
plt.xlabel('Meses')
plt.ylabel('Total de personas confirmadas')
plt.legend()

plt.show()
```

```python
#Estilo orientado a objetos

fig, ax = plt.subplots(figsize=(8,5), dpi=100)

ax.bar(x_valores, val_df,color = 'b', width = 0.2, label='DF')
ax.bar(x_valores + 0.25, val_jal, color = 'r', width = 0.2, label='JAL')
ax.bar(x_valores + 0.5, val_yuc, color = 'c', width = 0.2, label='YUC')

ax.set_xticks(x_valores)
ax.set_xticklabels(meses)
ax.legend()
ax.set(xlabel= 'Meses' , ylabel='Total de personas confirmadas')

plt.show()
```
